
# Data Analysis and Visualizations

```python
# how many people was cancelled and not cancelled booking, we show this in % format using normalize =  True in value_counts
# 0 represent not cancelled and 1 represent cancelled
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
sns.set(rc = {'figure.figsize':(15,5)})
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.6)
plt.show() # using matplotlib libraries
```

We clearly saw that 62% people was not cancelled booking but 37% people do

```python
# reservation status in different hotels
ax = sns.countplot(data =  df,x = 'hotel',hue = 'is_canceled')
plt.title('reservation status in different hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of observation')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
# calculating resort and city hotel cancellation in percentage
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

We clearly see that not cancellation rate is higher in Resort Hotels while cancellation rate is higher in City Hotel

```python
# we check is price affect on hotel cancellation or not {in both hotels}
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
sns.set(rc = {'figure.figsize':(15,5)})
plt.title("Average Dairy Rate in resort and city hotel")
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')

plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 15)
plt.show()
```

```python
# reservation status per month
# Convert the 'reservation_status_date' column to datetime explicitly
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

# Extract the month from the 'reservation_status_date' column
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize = (15,5))
ax = sns.countplot(x = 'month',hue = 'is_canceled',data = df)
plt.title('Reservation status per month')
plt.xlabel('Month')
plt.ylabel('Number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
# Average dairy rate per month 
plt.figure(figsize = (15,6))
plt.title("ADR per month",fontsize = 20)
# Filter the DataFrame and calculate the sum of 'adr' by month for canceled bookings
df_canceled = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Create the bar plot
sns.barplot(x='month', y='adr', data=df_canceled)
plt.show()

plt.show()
```

Afetr looking at month 1(Janeuary) we clearly prove that,If ADR is higher than cancellation rate is also higher so 

```python
# Cancellation rate per country
cancelled_data = df[df['is_canceled'] ==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (15,6))
plt.title("Top 10 country with reservation cancellation")
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```

From above chart we saw thar PRT(portugese) has higher cancellation rate as compare to others and on second spot it's a GBR
